### 1. What is a Support Vector Machine (SVM)?
**Answer:** A Support Vector Machine (SVM) is a supervised learning algorithm that finds the optimal hyperplane separating classes by maximizing the margin between them.

### 2. What is the difference between Hard Margin and Soft Margin SVM?
**Answer:** Hard margin assumes perfectly separable data, while Soft margin allows some misclassification controlled by parameter C.

### 3. What is the mathematical intuition behind SVM?
**Answer:** SVM maximizes the margin between classes while minimizing classification error using convex optimization.

### 4. What is the role of Lagrange Multipliers in SVM?
**Answer:** They convert constrained optimization into the dual optimization problem enabling efficient computation.

### 5. What are Support Vectors in SVM?
**Answer:** Support vectors are the data points closest to the decision boundary that determine the hyperplane.

### 6. What is a Support Vector Classifier (SVC)?
**Answer:** SVC is the classification implementation of SVM.

### 7. What is a Support Vector Regressor (SVR)?
**Answer:** SVR is the regression implementation of SVM used for predicting continuous values.

### 8. What is the Kernel Trick in SVM?
**Answer:** The kernel trick allows nonlinear classification by implicitly mapping features into higher-dimensional space.

### 9. Compare Linear, Polynomial, and RBF kernels.
**Answer:** Linear kernel creates straight boundaries, Polynomial kernel produces polynomial curves, and RBF kernel produces flexible nonlinear decision boundaries.

### 10. What is the effect of the C parameter in SVM?
**Answer:** C controls the trade-off between maximizing margin and minimizing classification errors.

### 11. What is the role of Gamma in RBF kernel?
**Answer:** Gamma determines the influence of a single training example; higher gamma means smaller influence radius.

### 12. What is the Naïve Bayes classifier and why is it called Naïve?
**Answer:** It is a probabilistic classifier assuming conditional independence between features, hence 'naïve'.

### 13. What is Bayes’ Theorem?
**Answer:** P(A|B) = P(B|A)P(A)/P(B).

### 14. Differences between Gaussian, Multinomial, and Bernoulli NB?
**Answer:** Gaussian handles continuous data, Multinomial handles count data, and Bernoulli handles binary features.

### 15. When should you use Gaussian Naïve Bayes?
**Answer:** When features are continuous and approximately normally distributed.

### 16. Key assumptions made by Naïve Bayes?
**Answer:** Features are conditionally independent and contribute equally.

### 17. Advantages and disadvantages of Naïve Bayes?
**Answer:** Advantages: fast and simple. Disadvantages: independence assumption unrealistic.

### 18. Why is Naïve Bayes good for text classification?
**Answer:** It performs well with high‑dimensional sparse word features.

### 19. Compare SVM and Naïve Bayes for classification.
**Answer:** SVM handles complex decision boundaries better, while Naïve Bayes is faster and simpler.

### 20. How does Laplace smoothing help in Naïve Bayes?
**Answer:** It prevents zero probabilities by adding a small constant to counts.
